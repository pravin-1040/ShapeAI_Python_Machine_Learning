```python
import numpy as np
import pandas as pd
import sklearn
```

Now We Will load the data that we will be using to train our model

```python
from sklearn.datasets import load_boston
df = load_boston()
```

```python
df.keys()         #Returns all the keys of the dataset dictionaries
```

```python
print(df.DESCR)     #Info about the dataset
```

```python
print(df.feature_names) 
```

```python
print(df.data) 
```

We convert our dataset into the pandas dataframe,so that it is easier to analyze the data

```python
boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head(10)
```

Adding a new column to the target values of dataframe

```python
boston['MEDV'] = df.target
boston.head()
```

Check if the dataset contain any null value or not

```python
boston.isnull()
```

```python
boston.isnull().sum()
```

We never train the model on all the data that we have, we always make sure to at least have a test dataset, which is different fron the training dataset.

```python
from sklearn.model_selection import train_test_split

x = boston.drop('MEDV', axis = 1)
y = boston['MEDV']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=5)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

```

Now lets import the Linear Regression Model from sklearn and trai it on the training dataset.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

```

```python
## Fitting model on training dataset
lin_model = LinearRegression()

lin_model.fit(x_train, y_train)
```

```python
y_train_predict = lin_model.predict(x_train)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))

print("The model perfomance for testing set")
print('RMSE is {}'.format(rmse))
print('\n')

##on testing set
y_test_predict = lin_model.predict(x_test)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))

print("The model perfomance for testing set")
print('RMSE is {}'.format(rmse))
print('\n')
```
